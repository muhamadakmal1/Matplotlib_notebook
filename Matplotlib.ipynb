we can make many plot on one page

```python
plt.figure()
plt.subplot(2,2,1)    #<--- use for part of page(divide the page)
#1s plot
plt.plot(x, np.sin(x))

# 2nd plot
plt.subplot(2,2,2)
plt.plot(x, np.cos(x))

# 3rd plot
plt.subplot(2,2,3)
plt.plot(x, np.tan(x))

# 4th plot
plt.subplot(2,2,4)
plt.plot(x, np.cos(x+1))

plt.show()
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.plot(x, np.sin(x-1), color = 'g')
plt.plot(x, np.sin(x-2), color = (1.0, 0.5, 0)) #rgb
plt.plot(x, np.cos((x)), color = ((0.1, 0.2, 0.5, 1)))
```

**We can change limits of x and y axis**

```python
plt.plot(x, np.sin(x), color = 'blue')
plt.ylim(-1.5, 2.5)
plt.xlim(1, 15)
plt.show()
```

```python
plt.plot(x, np.sin(x), color = 'blue')
plt.ylim(-1.5, 2.5)
plt.xlim(1, 15)
plt.axis('equal')  # use to making beautiful plot
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid()
plt.axis('equal')

plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid()
plt.title("This is sin plot before real dataset execution to understand basic syntax")
plt.axis('equal')
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid()
plt.title("This is sin plot before real dataset execution to understand basic syntax")
plt.xlabel("x-axis") # label name x axis
plt.ylabel("x-axis")  # label name y axis
plt.axis('equal')
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue', label = 'Sinx')
plt.plot(x, np.cos(x-0), color = 'red', label = 'Cosx')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid()
plt.title("This is sin plot and cos before real dataset execution to understand basic syntax")
plt.xlabel("x axis")
plt.ylabel("y axis")

#plt.axis('equal')

plt.legend()  # helping information of label on plot.
plt.show()
```

```python
# ------Histogram---------h
#  A histogram is a graph that shows the distribution of numerical data.

data = np.random.randn(1000) #10 bins
plt.hist(data, bins = 20, color ='Red')   # bin is smallest unit to divide the data. by default bins = 10,
plt.show()
```

```python
data = np.random.randn(1000) #bin =10

plt.hist(data, bins = 20, histtype = 'step', color = 'green')
plt.show()
```

```python
# 3d Plot

x = np.random.randn(50)
y = np.random.randn(50)
z = np.random.randn(50)

fig = plt.figure()

ax = fig.add_subplot(projection = '3d')

ax.scatter(x,y,z, c='#FF5733', alpha=1)

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
# pie bar

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```
